Importing numpy, pyplot and pandas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Reading the dataset

```python
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
```

Train test split

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,stratify=y,random_state=40)
```

Scaling

```python
#from sklearn.preprocessing import StandardScaler
#scaler=StandardScaler()
#x_train=scaler.fit_transform(x_train)
#x_test=scaler.transform(x_test)
```

Building Model

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_graphviz
model=DecisionTreeClassifier(max_depth=4)
model.fit(x_train,y_train)
```

Drawing the figure

```python
plt.figure("Decision Tree Graph", figsize=[12, 6])
plot_tree(model, feature_names=iris.feature_names, class_names=iris.target_names.tolist())
plt.tight_layout()
plt.show()

```

```python
# prediction
y_pred=model.predict(x_test)
y_pred_prob=model.predict_proba(x_test)

```

```python
y_pred
```

Evaluate

```python
labels_order=[0,1,2]
# f1
from sklearn.metrics import f1_score
f1=f1_score(y_true=y_test,y_pred=y_pred,labels=labels_order,average="weighted")
print(f"f1: {f1}")
```

```python
# accuracy
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_true=y_test, y_pred=y_pred)
print(f"Accuracy: {acc}")
```
